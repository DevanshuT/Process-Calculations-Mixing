# Hello ! This is a program to do Chemical process calculations

```python
import ipywidgets as ipw
from IPython.display import display
```

```python
MixingButton = ipw.Button(
    description='Mixing Process',
    disabled=False,
    button_style='info', # 'success', 'info', 'warning', 'danger' or ''
    tooltip='Click me',
    icon='',
)

ipw.Dropdown(
    options=['2', '3'],
    value='2',
    description='Number:',
    disabled=False
    )


stream1 = ipw.Text(
        value='',
        placeholder='Enter total mass of stream1',
        description='Stream1',
        disabled=False
        )


stream2 = ipw.Text(
        value='',
        placeholder='Enter total mass of stream2',
        description='Stream2',
        disabled=False
        )

stream3 = ipw.Text(
        value='',
        placeholder='Enter total mass of stream3',
        description='Stream3',
        disabled=False
        )

file = open("Mixing.png", "rb")
image = file.read()
Mixing_diagram = ipw.Image(
    value=image,
    format='png',
    width=300,
    height=400,
    )
```

```python

def Mixing(self):
    print("This is a program to calculate streams and componenets for mixing process")
    print("Here is the diagram")
    display(Mixing_diagram)
    display(stream1)
    display(stream2)
    display(stream3)
    


MixingButton.on_click(Mixing)
MixingButton

```
